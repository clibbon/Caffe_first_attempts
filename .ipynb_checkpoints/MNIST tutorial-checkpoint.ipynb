This is my first attempt at using CAFFE. This follows the tutorial http://caffe.berkeleyvision.org/gathered/examples/mnist.html

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

caffe_root = '/home/alex/Caffe/caffe/'
import sys
sys.path.insert(0, caffe_root + 'python')

import caffe

# Set paths to the model definition file, model weights, 
# and image to be classified
MODEL_FILE = caffe_root + 'models/bvlc_reference_caffenet/deploy.prototxt'
PRETRAINED = caffe_root + 'models/bvlc_reference_caffenet/bvlc_reference_caffenet.caffemodel'
IMAGE_FILE = caffe_root + 'images/cat.jpg'

# Checking that the file exists
import os
if not os.path.isfile(PRETRAINED):
    print("Pre-trained file is missing")
else:
    print("All successfully loaded")
```

This section sets the classifier

```python
caffe.set_mode_cpu()
net = caffe.Classifier(MODEL_FILE, PRETRAINED,
                       mean=np.load(caffe_root + 'python/caffe/imagenet/ilsvrc_2012_mean.npy').mean(1).mean(1),
                       channel_swap=(2,1,0),
                       raw_scale=255,
                       image_dims=(256, 256))
```

I wish my kitty were as cute as the one in the demo but this is the best I could find

```python
input_image = caffe.io.load_image(IMAGE_FILE)
plt.imshow(input_image)
```

And now hopefully to move onto the classification. Default classifier does 10 predictions - see tutorial for details

```python
prediction = net.predict([input_image])
print 'prediction shape:', prediction[0].shape
plt.plot(prediction[0])
print 'predicted class:', prediction[0].argmax()
```

A few helper functions quickly defined

```python
# Finds the name of the identified class
def whatsThat(pred_class):
    print type(pred_class)
    if type(pred_class) != 'list':
        pred_class = [pred_class]
    fp = open(caffe_root + "data/ilsvrc12/synset_words.txt")
    for i, line in enumerate(fp):
        print pred_class
        if i in pred_class:
            print line
    fp.close()
    return True
```
